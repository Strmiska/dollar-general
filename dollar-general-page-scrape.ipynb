```python
import re
import requests
import pandas as pd

from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# import url list
url_df = pd.read_csv('dollarGeneralUrls.csv')
```

```python
url_df.head()
```

```python
# bring the pages to max capacity...
for i in url_df.urls:
    print i + "?product_list_limit=96"
```

```python
PHANTOMJS_PATH = './phantomjs'
browser = webdriver.PhantomJS(PHANTOMJS_PATH)
```

```python
# Loop Through Database...
itemsList = []

for url in url_df.urls:
    
    nextPage = True
    pgcount = 1
    
    while nextPage == True:
        
        browser.get(url + "?p=" + str(pgcount) + "&product_list_limit=96")
        soup = BeautifulSoup(browser.page_source, "html.parser")
        
        #### gather items ####

        for row in soup.find_all('div', {"class": "product details product-item-details"}):
            itemsDic = {}

            # get department
            if re.search('.com\/(\S+)\/', url) == None:
                itemsDic['dept'] = 'nan'
            else:
                itemsDic['dept'] = re.search('.com\/(\S+)\/', url).group(1)

            # get product name
            if row.find('a') == None:
                itemsDic['name'] = 'nan'
            else:
                itemsDic['name'] = row.find('a').text.strip()

            # get SKU
            if row.find('strong', {'class': 'product-item-sku-qv'}) == None:
                itemsDic['sku'] = 'nan'
            else:
                itemsDic['sku'] = row.find('strong', {'class': 'product-item-sku-qv'}).text.strip()

            # get price
            if len(soup.find('div', {"class": "product details product-item-details"})) > 1:
                tinylist = []
                for price in soup.find('div', {"class": "product details product-item-details"}).find_all('span', {'class': 'price'}):
                    tinylist.append(price.text.strip())
                itemsDic['price'] = tinylist
            else:
                itemsDic['price'] = soup.find('div', {"class": "product details product-item-details"}).find('span', {'class': 'price'}).text.strip()


            itemsList.append(itemsDic)

        #### gather items ####

        print url
        print "Gathering items for page: " + str(pgcount)

        if soup.find('a', {"class": "action next"}) == None:
            nextPage = False
        else:
            nextPage = True
            pgcount+=1
        
        
```

```python
general_df = pd.DataFrame(itemsList)
```
